### Удалите ненужные колонки и недостающие значения.

#### Я не сильно разбираюсь в футболе, и думаю при реальной работе с подобным датафреймом рядом был бы специалист который знает больше что-то вроде скаута который понимает основные характеристики и во вторых нам не поставленна конкретная задача, что именно мы будем анализировать, поэтому удаление ненужных столбцов я сделаю на свое усмотрение. И так удалим следующее

- Preferred Foot
- Contract Valid Until	
- Joined
- Name так как есть ID (да там есть пара дубликатов но это не критично)

```python
df_result_1 = df.drop(['Preferred Foot', 'Contract Valid Until', 'Joined', 'Name'], axis=1)
df_result_1.info()
```

Так как нет описания дата-сета, сложно понять что озанчают некоторые из колонок, но нам ясно видно что очень много нулей в столбце с клубом, и если их мы можем заполнить значением к приеру "свободный агент" то непонятно что означающее валуе, можно заполнить средним, но проще все удалить, к тому же помойму именно это и требуется в задании

```python
df_result_1 = df_result_1.dropna()
```

```python
df_result_1.isnull().sum()
```

```python
df_result_1.info()
```

### Найдите в датафрейме полные дубликаты и удалите их. Значения могут быть одинаковыми, но написаны по-разному. Например, может отличаться размер регистра (заглавные и строчные буквы). Особое внимание уделить колонке с названиями команд.

Так как в колонке с названиями Club, Nationality, Position у нас в принципе не может отсутвовать дубли, то их мы попросту трогать не будем

```python
list_columns = list(df_result_1)
list_columns.remove('Club')
list_columns.remove('Nationality')
list_columns.remove('Position')
```

так как оставшиеся объекты имеют числовой тип, просто используем drop_duplicates

```python
df_result_2 = df_result_1.drop_duplicates(subset=list_columns)
```

```python
df_result_2.info()
```

### Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36. 

```python
cut_labels_4 = ['до 20', 'от 20 до 30', 'от 30 до 36', 'старше 36']
cut_bins = [0, 19, 29, 35, 100]
df_result_2['Age group'] = pd.cut(df_result_2['Age'], 
                                 bins=cut_bins,
                                 labels=cut_labels_4)
```

```python
df_result_2
```
